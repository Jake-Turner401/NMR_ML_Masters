```python
#Testing all parameters

import utils 
import sklearn
from sklearn.metrics import mean_squared_error
import os
import shutil
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Finding the optimal r_cut
r_cut = range(20,75,5)
inp = 'rat.xyz'
spec = ['O','Si']
lm = [0,1,2]

utils.env_configure()

if not os.path.exists('r_cut'):
    os.mkdir('r_cut')

for i in r_cut:
    i = i/10
    print(i)
    for l in spec:
        print(l)
        utils.get_ps_param(inp, lm, 400, i)
        utils.env_sparse(lm, 1000)
        utils.apply_env_sparse(lm)
        utils.kerneller(2)
        utils.model_trainer(l, 0.9, '1e-8')
        
        if not os.path.exists('r_cut/' +str(i) + l):
            os.mkdir('r_cut/' +str(i) + l)
        
        shutil.move('prediction_cartesian.txt','r_cut/' +str(i) + l +'/' )
        shutil.move('prediction_L0.txt','r_cut/' +str(i) + l +'/' )
        shutil.move('prediction_L1.txt','r_cut/' +str(i) + l +'/' )
        shutil.move('prediction_L2.txt','r_cut/' +str(i) + l +'/' )
        
        
```
