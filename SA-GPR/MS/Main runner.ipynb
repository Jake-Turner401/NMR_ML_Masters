```python
#Run model for range of paramaters

import utils 
import sklearn
from sklearn.metrics import mean_squared_error
import os
import shutil
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
#tr_size

tr_sizes = [1000,2000,3000,4000,5000,6000]
cat=['Rattle']
inp = ['rat.xyz']
spec = ['O','Si']
lm = [0,1,2]

utils.env_configure()


if not os.path.exists('tr_size'):
    os.mkdir('tr_size')

for t in cat:
    if not os.path.exists('tr_size/' + t):
        os.mkdir('tr_size/' + t)
    ind = cat.index(t)
    counter = 0
     
    utils.get_ps_param(inp[ind], lm, 400, 3.5)
    utils.env_sparse(lm, 1000)
    utils.apply_env_sparse(lm)  
    
    while counter < 5:    
        for i in tr_sizes:
            print(i)
            for l in spec:
                print(l)
                if l == 'O':
                    tr_size = int(i*2/3)
                elif l == 'Si':
                    tr_size = int(i/3)

                utils.kerneller(2)
                utils.model_trainer(l, tr_size, '1e-8')

                if not os.path.exists('tr_size/' + t + '/' +str(i) + l + str(counter)):
                    os.mkdir('tr_size/' + t + '/' +str(i) + l + str(counter))


                shutil.move('prediction_cartesian.txt','tr_size/' + t + '/' +str(i) + l + str(counter) +'/' )
                shutil.move('prediction_L0.txt','tr_size/' + t + '/' +str(i) + l + str(counter) +'/' )
                shutil.move('prediction_L1.txt','tr_size/' + t + '/' +str(i) + l + str(counter) +'/' )
                shutil.move('prediction_L2.txt','tr_size/' + t + '/' +str(i) + l + str(counter) +'/' )
        counter +=1
```
